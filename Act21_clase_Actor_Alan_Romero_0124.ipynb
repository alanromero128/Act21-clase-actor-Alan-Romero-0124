### Nombre de clase :  ACTOR

```python
class Actor:
    def __init__(self, nombre, telefono, edad):
        # Atributos de la clase
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    # Funciones (Métodos)
    def llorar(self):
        print(f"{self.nombre} está llorando con mucha emoción... ¡Qué actuación!")

    def ensayar(self):
        print(f"{self.nombre} está repasando sus líneas para la próxima escena.")

    def comprender_personaje(self):
        print(f"{self.nombre} está analizando las motivaciones de su personaje.")
# --- Creación de Objetos (Instancias) ---
# Objeto 1 con datos de ejemplo
actor1 = Actor("Gael García", "555-0123", 45)

# Objeto 2 con datos de ejemplo
actor2 = Actor("Salma Hayek", "555-9876", 57)
# --- Uso de los objetos y sus funciones ---
print(f"--- Ficha del Actor 1 ---")
print(f"Nombre: {actor1.nombre}\nTeléfono: {actor1.telefono}\nEdad: {actor1.edad}")
actor1.ensayar()
actor1.comprender_personaje()
actor1.llorar()

print("\n" + "-"*25 + "\n")

print(f"--- Ficha del Actor 2 ---")
print(f"Nombre: {actor2.nombre}\nTeléfono: {actor2.telefono}\nEdad: {actor2.edad}")
actor2.ensayar()
actor2.comprender_personaje()
actor2.llorar()

print("Programa terminado ")
print("programa realizado por Alan Romero NL = 47")
```
